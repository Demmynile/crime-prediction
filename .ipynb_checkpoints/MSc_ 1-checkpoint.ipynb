```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

# Load the datasets
leeds_theft_df = pd.read_csv('leeds_theft_data.csv')
join_df_poi_lsoa = gpd.read_file('./Merged/join_df_poi_lsoa.shp')

# Create geometry column from latitude and longitude in leeds_theft_df
leeds_theft_df['geometry'] = leeds_theft_df.apply(lambda row: Point(row['Longitude'], row['Latitude']), axis=1)

# Convert to GeoDataFrame
leeds_theft_gdf = gpd.GeoDataFrame(leeds_theft_df, geometry='geometry')

# Set CRS to WGS84 (EPSG:4326) for leeds_theft_gdf
leeds_theft_gdf.set_crs(epsg=4326, inplace=True)

# Check and convert CRS for join_df_poi_lsoa to WGS84 (EPSG:4326) if necessary
if join_df_poi_lsoa.crs != 'EPSG:4326':
    join_df_poi_lsoa = join_df_poi_lsoa.to_crs(epsg=4326)

# Perform the spatial join with specified suffixes for overlapping column names
result = gpd.sjoin(leeds_theft_gdf, join_df_poi_lsoa, how='inner', predicate='intersects', lsuffix='left', rsuffix='right')

# Display the first few rows of the result
print(result.head())
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

# Load the datasets
leeds_theft_df = pd.read_csv('leeds_theft_data.csv')
join_df_poi_lsoa = gpd.read_file('./Merged/join_df_poi_lsoa.shp')

# Create geometry column from latitude and longitude in leeds_theft_df
leeds_theft_df['geometry'] = leeds_theft_df.apply(lambda row: Point(row['Longitude'], row['Latitude']), axis=1)

# Convert to GeoDataFrame
leeds_theft_gdf = gpd.GeoDataFrame(leeds_theft_df, geometry='geometry')

# Set CRS to WGS84 (EPSG:4326) for leeds_theft_gdf
leeds_theft_gdf.set_crs(epsg=4326, inplace=True)

# Check CRS for join_df_poi_lsoa and convert if necessary
if join_df_poi_lsoa.crs != 'EPSG:4326':
    join_df_poi_lsoa = join_df_poi_lsoa.to_crs(epsg=4326)

# Buffer points in join_df_poi_lsoa to create small polygons
join_df_poi_lsoa['geometry'] = join_df_poi_lsoa.geometry.buffer(0.0001)

# Perform the spatial join with polygons from leeds_theft_gdf and buffered points from join_df_poi_lsoa
result = gpd.sjoin(leeds_theft_gdf, join_df_poi_lsoa, how='inner', op='intersects')

# Display the first few rows of the result
print(result.head())
```

#### 2. Joining the OSM, Boundary Data GB, DIGIMap and LSOA shape files data for Leeds

```python
# Loading all the shapefiles
osm_leeds = gpd.read_file('./Leeds.osm.shp/Leeds-shp/shape')
leeds_grid = gpd.read_file('./Grid/gb-grids')
boundary_gb = gpd.read_file('./Great_Britain_GRID_shapefile/gb_1km.shp')

# Ensuring all dataframes use the same CRS
osm_leeds = osm_leeds.to_crs(leeds_grid.crs)
boundary_gb = boundary_gb.to_crs(leeds_grid.crs)

# Dropping 'index_left' and 'index_right' columns 
for df in [osm_leeds, leeds_grid, boundary_gb]:
    if 'index_left' in df.columns:
        df.drop(columns=['index_left'], inplace=True)
    if 'index_right' in df.columns:
        df.drop(columns=['index_right'], inplace=True)

# Performing spatial joining between osm_leeds and leeds_grid
leeds_osm_grid_join = gpd.sjoin(osm_leeds, leeds_grid, how='inner', op='intersects')

# Dropping 'index_left' and 'index_right' columns in the joined result
if 'index_left' in leeds_osm_grid_join.columns:
    leeds_osm_grid_join.drop(columns=['index_left'], inplace=True)
if 'index_right' in leeds_osm_grid_join.columns:
    leeds_osm_grid_join.drop(columns=['index_right'], inplace=True)

# Performing spatial joining between the result and boundary_gb
joined_leeds_shp = gpd.sjoin(leeds_osm_grid_join, boundary_gb, how='inner', op='intersects')

# Saving to a new shapefile
joined_leeds_shp.to_file("joined_leeds.shp")

# Output the result
joined_leeds_shp
```
