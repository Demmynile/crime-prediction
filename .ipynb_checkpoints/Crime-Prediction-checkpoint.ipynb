### Loop through to combine all Police data together

```python
# write a function that can loop through the csv files in a folder with the same data format and concatenate them together.

def concatenate_csv_files_in_subfolders(main_folder_path):
   
    df_list = []

    for root, _, files in os.walk(main_folder_path):
        for file in files:
            if file.endswith('.csv'):
                file_path = os.path.join(root, file)
                df = pd.read_csv(file_path)
                df_list.append(df)

    concatenated_df = pd.concat(df_list, ignore_index=True)
    return concatenated_df

# Example usage
main_folder_path = './data'
police_df = concatenate_csv_files_in_subfolders(main_folder_path)
police_df
```

### Check all the plots available

```python
import glob
```

```python
# Define the folder containing the shapefiles
folder_path = './data/hull-shapefile'

# Find all shapefiles in the folder
shapefiles = glob.glob(os.path.join(folder_path, '*.shp'))

# Load each shapefile into a GeoDataFrame and check for empty geometries
gdfs = []
for shp in shapefiles:
    gdf = gpd.read_file(shp)
    if not gdf.empty and gdf.geometry.notnull().all():
        gdfs.append(gdf)

# Ensure all GeoDataFrames have the same CRS
if gdfs:
    common_crs = gdfs[0].crs
    gdfs = [gdf.to_crs(common_crs) for gdf in gdfs]

    # Visualize the geometries
    for i, gdf in enumerate(gdfs):
        gdf.plot()
        plt.title(f'Shapefile {i}')
        plt.show()
```

### Filtering The Boundary for Hull City from the Boundary data

```python
from shapely.geometry import box

# Step 1: Read the Great Britain GRID shapefile
grids = osm_boundary_1km

# Step 2: Define a bounding box for Hull City (adjust coordinates as needed)
hull_bbox_coords = [507000, 427000, 517000, 437000]  # [minx, miny, maxx, maxy]
hull_bbox = box(*hull_bbox_coords)
hull_bbox_gdf = gpd.GeoDataFrame({'geometry': [hull_bbox]}, crs="EPSG:27700")

# Step 3: Ensure the grids GeoDataFrame has the same CRS as the Hull bounding box
grids = grids.to_crs(hull_bbox_gdf.crs)

# Step 4: Perform the spatial join to filter grids that intersect with Hull City's bounding box
hull_grids = gpd.sjoin(grids, hull_bbox_gdf, how="inner", op="intersects")

# Step 5: Drop the extra columns added by the spatial join
hull_grids = hull_grids.drop(columns=['index_right'])

# Step 6: Save the filtered data to a new shapefile
hull_grids.to_file('path_to_save_hull_grids.shp')
```

### Merging both boundary data , digimap data and osm data

```python
# Writing a code that would measure all

target_crs = 'EPSG:4326'

hull_grid_data = hull_grid_data.to_crs(target_crs)
digimap_data_hull_df = digimap_data_hull_df.to_crs(target_crs)
osm_df = osm_df.to_crs(target_crs)
lsoa_hull = lsoa_hull.to_crs(target_crs)
# Join digimap data and  boundary data
digi_grid_data = gpd.sjoin(hull_grid_data, digimap_data_hull_df, how="inner", op="intersects")

# Drop the extra columns added by the spatial join
digi_grid_data = digi_grid_data.drop(columns=['index_right'])

# Join the result with osm data
digi_grid_osm = gpd.sjoin(osm_df, digi_grid_data, how="inner", op="intersects")

# Drop the extra columns added by the spatial join
digi_grid_osm = digi_grid_osm.drop(columns=['index_right'])

# Join the result with osm data
digi_grid_osm_lsoa = gpd.sjoin(digi_grid_osm, lsoa_hull  , how="inner", op="intersects")

# Drop the extra columns added by the spatial join
digi_grid_osm_lsoa = digi_grid_osm_lsoa.drop(columns=['index_right'])

digi_grid_osm_lsoa
```
