### 1.0 Acquiring the Independent data

```python
# importing libraries
import pandas as pd
import geopandas as gpd

# remove warnings
import warnings
warnings.filterwarnings("ignore")
```

#### 1.1 LSOA_HULL 

```python
# Acquiring the lsoa data exported in the first notebook as a shapefile
lsoa_hull = gpd.read_file("lsoa_hull.shp")
lsoa_hull.head()
```

#### 1.2 CENSUS DATA

```python
# write a function to filter out anything that contains  hull in the LSOA
def filterHull(data):
    data = data[data['Lower layer Super Output Areas'].str.contains('Hull')]
    return data
```

```python
# acquiring the first census data , run the function to filter then resample due to too much data and visualize
df_census_first = pd.read_csv('./data/poi_4_all/custom-filtered-2024-06-20T09_12_05Z.csv')
df_census_first = filterHull(df_census_first)
df_census_first = df_census_first.sample(frac=0.04, random_state=None)
df_census_first
```

```python
# acquiring the second census data , then resample due to too much data and visualize
df_census_second = pd.read_csv('./data/poi_4_all/custom-filtered-2024-08-27T18_06_21Z.csv')
df_census_second = df_census_second.sample(frac=0.04, random_state=None)
df_census_second
```

```python
# acquiring the third census data , then resample due to too much data and visualize
df_census_third = pd.read_csv('./data/poi_4_all/custom-filtered-2024-08-27T18_11_02Z.csv')
df_census_third = df_census_third.sample(frac=0.04, random_state=None)
df_census_third
```

```python
# acquiring the fourth census data , then resample due to too much data and visualize
df_census_fourth = pd.read_csv('./data/poi_4_all/custom-filtered-2024-08-27T18_14_34Z.csv')
df_census_fourth = df_census_fourth.sample(frac=0.04, random_state=None)
df_census_fourth
```

```python
# acquiring the fifth census data , then resample due to too much data and visualize
df_census_fifth = pd.read_csv('./data/poi_4_all/custom-filtered-2024-08-27T18_18_50Z.csv')
df_census_fifth = df_census_fifth.sample(frac=0.04, random_state=None)
df_census_fifth
```

```python
# Merging all the census dataframe together with a key as a join
census_df = pd.merge(df_census_first, df_census_second, on='Lower layer Super Output Areas Code', how='outer')
census_df = pd.merge(census_df, df_census_third, on='Lower layer Super Output Areas Code', how='outer', suffixes=('_df2', '_df3'))
census_df = pd.merge(census_df, df_census_fourth, on='Lower layer Super Output Areas Code', how='outer', suffixes=('_df3', '_df4'))
census_df = pd.merge(census_df, df_census_fifth, on='Lower layer Super Output Areas Code', how='outer', suffixes=('_df4', '_df5'))
census_df
```

```python
# Keep the first occurrence of each column name and drop duplicates
census_df = census_df.loc[:, ~census_df.columns.duplicated()]
census_df
```

```python
# Check the data types
census_df.dtypes
```

```python
# Check for missing values across all the columns
census_df.isnull().sum()
```

```python
# dropping some unwanted columns in the dataframe
census_df.drop(["Lower layer Super Output Areas_x" , "Lower layer Super Output Areas_y" , "Lower layer Super Output Areas_df3" , "Lower layer Super Output Areas_df4"] , axis = 1 , inplace = True) 
```

```python
# Check for missing values across all the columns
census_df.isnull().sum()
```

```python
# dropping all na values
census_df = census_df.dropna()
census_df
```

```python
# exporting the census dataframe as census_df
census_df.to_csv("census_df")
```

#### 1.3 MERGING CENSUS_DF WITH LSOA

```python
#renaming the lsoa code with Lower layer Super Output Areas Code
lsoa_hull.rename({'LSOA code' : 'Lower layer Super Output Areas Code'} , axis = 1 , inplace = True)
```

```python
# Merging the census data with lsoa data
GDF_LSOA_CENSUS = lsoa_hull.merge(census_df , on = 'Lower layer Super Output Areas Code')
GDF_LSOA_CENSUS = GDF_LSOA_CENSUS.sample(frac=0.8, random_state=None)
GDF_LSOA_CENSUS
```

```python
# checking if the merging has null values
GDF_LSOA_CENSUS.isnull().sum()
```

#### 1.4 GDF_GRID_500

```python
# acquiring the exported Geodataframe 500m grid data
GDF_GRID_500 = gpd.read_file('GDF_GRID_500.shp')

# Display the first five rows of the GeoDataFrame
GDF_GRID_500.head()
```

#### 1.5 MERGING GDF_LSOA_CENSUS X GDF_GRID_500

```python
# merging the GDF_LSOA_CENSUS with  GDF_GRID_500

GDF_GRID_500 = GDF_GRID_500.to_crs(epsg=27700)
GDF_LSOA_CENSUS = GDF_LSOA_CENSUS.to_crs(epsg=27700)

# Join digimap data and  boundary data
GDF_LSOA_CENSUS_500 = gpd.sjoin(GDF_GRID_500 , GDF_LSOA_CENSUS , how="inner", op="intersects")

# Drop the extra columns added by the spatial join
GDF_LSOA_CENSUS_500 = GDF_LSOA_CENSUS_500.drop(columns=['index_right'])

GDF_LSOA_CENSUS_500
```

```python
# checking the newly merged data with its data types
GDF_LSOA_CENSUS_500.dtypes
```

```python
# List the columns in their desired order
desired_order = ['grid_index', 'Crime ID', 'Month', 'Lower layer Super Output Areas Code']

# Create a list of remaining columns, excluding those already reordered
remaining_columns = [col for col in GDF_LSOA_CENSUS_500.columns if col not in desired_order]

# Combine the desired columns order with the remaining columns
new_column_order = desired_order + remaining_columns

# Reorder the GeoDataFrame based on the new column order
GDF_LSOA_CENSUS_500 = GDF_LSOA_CENSUS_500.reindex(columns=new_column_order)

# Display the first few rows to confirm the new order
GDF_LSOA_CENSUS_500
```

#### 1.6 OVERVIEW OF THE TOTALLY MERGED INDEPENDENT DATAFRAME

```python
# The independent variable that has been spatially analyzed is GDF_LSOA_CENSUS_500
x_data = GDF_LSOA_CENSUS_500
x_data
```

```python
# Converting the 'Month' column to datetime format
x_data['Month'] = pd.to_datetime(x_data['Month'], format='%Y-%m')
x_data.head()
```

```python
# checking for new arranged independent variable
x_data.dtypes
```

```python
# export the independent dataframe
x_data.to_csv("x_data.csv")
```
